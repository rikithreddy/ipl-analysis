```python
import glob
import os

import pandas as pd
from yaml import safe_load

if 'google.colab' in str(get_ipython()):
  from google.colab import drive
  drive.mount('/content/drive')
  
  BASE_PATH = 'drive/MyDrive/git_data/raw_data'
else:
  BASE_PATH = "./"

FILE_LIST = [os.path.join(BASE_PATH, f) for f in os.listdir(BASE_PATH)]

print(FILE_LIST[0])

match_info_df = pd.DataFrame()

for i, fname in enumerate(FILE_LIST):
  with open(fname, 'r') as f:
    data = safe_load(f)#["innings"][0]["1st innings"]["deliveries"]
    meta_data = pd.json_normalize(data)
    
    meta_data["match_id"] = i

    match_info_df = match_info_df.append(meta_data)
    # innings = pd.json_normalize(safe_load(f))

  if (i + 1) % 25 == 0:
    print("Proceesed {}".format(i))
```

```python
columns = [
           "city",
           "dates",
           "outcome_by_runs",
           "method",
           "winner",
           "overs",
           "player_of_match",
           "teams",
           "toss",
           "toss_winner",
           "umpires",
           "venue",
           "match_id",
           "outcome_by_wickets"
]

match_info_df = match_info_df.drop(columns = match_info_df.iloc[:, [0,1,2,3,5,7,8]].columns)
match_info_df.columns = columns
match_info_df = match_info_df.set_index("match_id")
match_info_df.head()
```
